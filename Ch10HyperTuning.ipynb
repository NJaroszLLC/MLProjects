```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
fashion_mn = keras.datasets.mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mn.load_data()
```

```python
X_valid, X_train = X_train_full[:5000]/255.0, X_train_full[5000:]/255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
def build_model(n_hidden=2, n_neurons=300, learning_rate=3e-1, input_shape=[28,28]):
    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation='relu'))
    model.add(keras.layers.Dense(10, activation="softmax"))
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=["accuracy"])
    return model
```

```python
keras_reg = keras.wrappers.scikit_learn.KerasClassifier(build_model)
history = keras_reg.fit(X_train, y_train, epochs=30,
             validation_data=(X_valid, y_valid),
             callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV
import sklearn
sklearn.__version__
```

```python
params_distribs = {
    "n_hidden" : [2,3,4],
    "n_neurons" : np.arange(300,400).tolist()
}

rnd_search_cv = RandomizedSearchCV(keras_reg, params_distribs, n_iter=5, scoring="accuracy", cv=3)
history = rnd_search_cv.fit(X_train, y_train, epochs=30,
                 validation_data=(X_valid, y_valid),
                 callbacks = [keras.callbacks.EarlyStopping(patience=7)])
 
```

```python
model = rnd_search_cv.best_estimator_.model
```

```python
rnd_search_cv.best_params_
```

```python
model.evaluate(X_test, y_test)
```
